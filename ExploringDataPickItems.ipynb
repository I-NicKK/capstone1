```python
import glob
import xmltodict
import json
from collections import defaultdict, Counter


def file_to_keep(file_name1, items_list):    
    file_name1_keep = []
    for file in file_name1:
        data = xmltodict.parse(open(file,'r').read())
        if isinstance(data['annotation']['object'], list):
            not_found = True
            for item in data['annotation']['object']:
                if not_found:
                    if item['name'] in items_list:
                        file_name1_keep.append(file)
                        not_found = False
        else:
            if data['annotation']['object']['name'] in items_list:
                file_name1_keep.append(file)
    return file_name1_keep


file_name1 = glob.glob('./data/1/*.xml')
file_name2 = glob.glob('./data/2/*.xml')
```

```python
count_items = Counter()
file_names_combined = [file_name1,file_name2]
for each_file_list in file_names_combined:
    for file in each_file_list:
        data = xmltodict.parse(open(file,'r').read())
        if isinstance(data['annotation']['object'], list): 
            for item in data['annotation']['object']:
                count_items[item['name']] += 1
        else:
            count_items[data['annotation']['object']['name']]
```
